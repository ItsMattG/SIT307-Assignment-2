```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/iris.txt', delimiter=',') 
```

```python
df.head()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.columns
```

```python
feature_1 = df.sepal_length.values
feature_2 = df.sepal_width.values
feature_3 = df.petal_length.values
feature_4 = df.petal_width.values
feature_5 = df['class'].values
```

```python
x = feature_1
y = feature_2

plt.figure(figsize=(14, 10))
plt.scatter(x, y, alpha=0.8, s=100)
plt.title("Sepal Comparsion")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
x = feature_3
y = feature_4

plt.figure(figsize=(14, 10))
plt.scatter(x, y, alpha=0.8, s=100)
plt.title("Petal Comparison")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```

```python
setosa=df[df['class']=='Iris-setosa']
versicolor =df[df['class']=='Iris-versicolor']
virginica =df[df['class']=='Iris-virginica']
```

```python
fig, ax= plt.subplots(1,2,figsize=(25, 11))

setosa.plot.scatter(x='sepal_length', y='sepal_width', color='red', label='setosa', ax=ax[0], s= 30)
versicolor.plot.scatter(x='sepal_length', y='sepal_width', color='green', label='versicolor', ax=ax[0], s= 30)
virginica.plot.scatter(x='sepal_length', y='sepal_width', color='blue', label='virginica', ax=ax[0], s= 30)

setosa.plot.scatter(x='petal_length', y='petal_width', color='red', label='setosa', ax=ax[1], s= 30)
versicolor.plot.scatter(x='petal_length', y='petal_width', color='green', label='versicolor', ax=ax[1], s= 30)
virginica.plot.scatter(x='petal_length', y='petal_width', color='blue', label='virginica', ax=ax[1], s= 30)

ax[0].set(title='Sepal Comparasion ', ylabel='sepal-width')
ax[1].set(title='Petal Comparasion',  ylabel='petal-width')
ax[0].legend()
ax[1].legend()
plt.show()
```

```python
# -- exploratory analysis -- #
# in scatterplot 1 & 2 the darker circles represent values that occur most often 
# in scatterplot 3 & 4 it's evident that setosa iris's have the largest sepal-width at the same time the lowest sepal length.
# additionally, the setosa has the smallest petal width & petal length.
# versicolor are sitting in the middle of the range for both lengths.
# finally, virginica iris's have the longest sepal and petal length as well as petal width
```
